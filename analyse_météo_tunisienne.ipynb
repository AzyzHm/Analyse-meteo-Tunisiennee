## Nettoyage des données:

Dans cette étape, nous avons supprimé toutes les lignes ayant au moins une valeur manquante ou au moins une valeur nulle et les lignes dupliquées. Nous avons aussi procédé à la suppression de colonnes ayant des valeurs erronées.

## Visualisation des données:

### Etude des Températures:

#### Variation Annuelle

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Année', y='Température_2m', data=mydata, marker='o', linestyle='-')
plt.xlabel('Année')
plt.ylabel('Température (°C)')
plt.title('Variation de la Température Moyenne au Fil des Années')
plt.grid(True)
plt.show()
```

- Ce graphique présente les fluctuations de la moyenne des températures maximales, moyennes et minimales au fil du temps. 

- Il est intéressant de noter que la température maximale a connu une tendance à la hausse depuis 1983, passant d'environ 19°C à près de 20°C en 1994, où elle est restée relativement stable jusqu'en 2004 avant de connaître une légère baisse à environ 19°C. Elle a ensuite suivi un cycle similaire jusqu'en 2016. Depuis 2016, sa valeur a augmenté pour atteindre un pic en 2022, approchant les 30°C.

- Les températures moyenne et minimale ont eu le même rythme où leurs valeurs respectives en 1983 étaient 18°C et prés de 17°C,  ont dépassées les 18°C et 19°C en 1994 pour chuter en 2004 avec des valeurs inférieures à 18°C et 17°C puis progressent jusqu'à 2016 pour atteindre les 20°C et 19°C en 2022.

#### Variation Saisonniére:

```python
# Copier les données originales pour ne pas les modifier
season_data = mydata.copy()

# Définir les saisons en fonction des mois
season_data['Saison'] = 'Hiver'
season_data.loc[(season_data['Mois'] >= 3) & (season_data['Mois'] <= 5), 'Saison'] = 'Printemps'
season_data.loc[(season_data['Mois'] >= 6) & (season_data['Mois'] <= 8), 'Saison'] = 'Été'
season_data.loc[(season_data['Mois'] >= 9) & (season_data['Mois'] <= 11), 'Saison'] = 'Automne'

# Calculer les statistiques saisonnières
seasonal_stats = season_data.groupby('Saison').agg({
    'Température_2m': ['mean', 'min', 'max'],
    'Vitesse_Vent_10m': ['mean', 'min', 'max'],
    'Précipitation_Totale_Corrigée': ['mean', 'sum']
})
print(seasonal_stats)
```

- Tout d'abord, il est clair que la température moyenne varie considérablement d'une saison à l'autre, avec des pointes atteignant 37.18°C pendant l'été et des baisses à 1.46°C pendant l'hiver. Cette variation témoigne des saisons distinctes que connaît le pays, avec des contrastes marqués entre les mois chauds et froids. 

- Parallèlement, la radiation solaire descendante suit également un schéma saisonnier, avec des niveaux plus élevés en été et en automne, suggérant une plus grande exposition au soleil pendant ces saisons. 

- En ce qui concerne la vitesse du vent, bien qu'elle reste relativement stable tout au long de l'année, des valeurs légèrement plus élevées sont observées pendant l'hiver. 

- Enfin, les précipitations totales corrigées montrent une certaine variabilité saisonnière, avec des moyennes annuelles les plus élevées en automne, suivies de près par le printemps. 

```python
# Graphique de variation saisonnière de la Température_2m
plt.figure(figsize=(12, 6))
sns.boxplot(x='Saison', y='Température_2m', data=season_data, order=['Hiver', 'Printemps', 'Été', 'Automne'])
plt.xlabel('Saison')
plt.ylabel('Température (°C)')
plt.title('Variation de la Température Moyenne par Saison')
plt.grid(True)
plt.show()
```

- Cette boite a mustache repésente **la variation de la temperature moyenne par saison** , voici un résumé des données:
    - **Hiver**: Les températures varient approximativement de 5 à 15°C, avec un point aberrant proche de 30°C.
    - **Printemps**: Les températures sont réparties entre environ 15 à 25°C, avec un point aberrant proche de 0°C.
    - **Été**: Les températures sont concentrées autour de 25 à 30°C.
    - **Automne**: Les températures varient entre environ 10 à presque 30°C, avec un point aberrant proche de 0°C.

### Analyse des Événements Météorologiques Extrêmes :

```python
# Identifier les événements météorologiques extrêmes (par exemple, vagues de chaleur)
heatwaves = mydata[mydata['Température_2m_Max'] > 35]  # Définir un seuil de température pour les vagues de chaleur

# Afficher les années et les mois des vagues de chaleur
print("Années et Mois des Vagues de Chaleur :")
print(heatwaves[['Année', 'Mois']].drop_duplicates().sort_values(by=['Année', 'Mois']))

#Distribution des Températures Maximales pendant les Vagues de Chaleur
plt.figure(figsize=(12, 6))
sns.histplot(data=heatwaves, x='Température_2m_Max', bins=30, color='yellow')
plt.xlabel('Température (°C)')
plt.title('Distribution des Températures Maximales pendant les Vagues de Chaleur')
plt.grid(True)
plt.show()

# Calculate the number of heatwaves per year
heatwaves_per_year = heatwaves.groupby('Année').size()

# Plot the number of heatwaves per year
plt.figure(figsize=(12, 6))
sns.lineplot(x=heatwaves_per_year.index, y=heatwaves_per_year.values, marker='o' , color='red')
plt.xlabel('Année')
plt.ylabel('Nombre de Vagues de Chaleur')
plt.title('Nombre de Vagues de Chaleur par Année')
plt.grid(True)
plt.show()
```

- À partir du 1er graphique, il est observé que le nombre de vagues de chaleur les plus significatives se situe entre 35°C et 38°C, avec une valeur cumulée d'environ 1262 vagues. Cependant, il est à noter qu'il peut y avoir des vagues de chaleur atteignant jusqu'à 46°C, bien que ces occurrences soient rares.

- D'après le 2e graphique, on observe que les années 1983 à 1991 ont connu entre 50 et 71 vagues de chaleur. En 1992, le nombre de vagues de chaleur a atteint un minimum, tombant en dessous de 30 vagues. Entre 1994 et 2020, le nombre de vagues varie principalement entre 50 et 84, avec un pic en 2022 où l'on compte environ 110 vagues de chaleur.

## Modèle de Prédiction de Température : Prévision des Tendances Climatiques Futures:

```python
# 1. Preprocessing
heatwaves['Année'] = heatwaves['Année'].astype(float)

# 2. Feature Selection
features = heatwaves[['Année']]
target = heatwaves['Température_2m_Max']

# Split the data into training set and test set
features_train, features_test, target_train, target_test = train_test_split(features, target, test_size=0.2, random_state=42)

# 3. Model Selection
model = LinearRegression()

# 4. Model Training
model.fit(features_train, target_train)

# Check the model performance on test data
target_pred = model.predict(features_test)
print('RMSE:', np.sqrt(mean_squared_error(target_test, target_pred)))

# 5. Prediction
future_years = np.array(range(2022, 2031)).reshape(-1, 1)
future_temps = model.predict(future_years)

# Create a DataFrame with the future years and temperatures
future_df = pd.DataFrame({'Année': future_years.flatten(), 'Température_2m_Max': future_temps})

# Append the future DataFrame to the original DataFrame
total_df = pd.concat([heatwaves, future_df])

# Plot the actual and predicted temperatures
plt.figure(figsize=(12, 6))
sns.lineplot(x='Année', y='Température_2m_Max', data=total_df,marker="o")
plt.xlabel('Année')
plt.ylabel('Température (°C)')
plt.title('Température Maximale Prédite pour 2022 à 2030')
plt.grid(True)
plt.show()
```

>Ce graphique suggère que la température maximale devrait atteindre près de 41°C en 2023, puis diminuer à 39°C, avant de progresser lentement jusqu'en 2030.
